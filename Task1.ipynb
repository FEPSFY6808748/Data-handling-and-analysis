```python
# relevant module imported
import math

# Array for numbers entered initialised
user_numbers = []

# Whilst the number of integers in the list user_numbers is lower than the index 5 (5 numbers indexed from 0), the entered 
# numbers are checked to be 4 digits in length using string handling and checked to not be in the user_numbers list to make
# sure they are unique. If either or both of these conditions are not met, then an error message is displayed and the user 
# is asked to enter a unique 4 digit integer again.
while len(user_numbers) < 5:
    user_int = int(input("Enter a 4-digit integer: "))
    if len(str(user_int)) == 4 and user_int not in user_numbers:
        user_numbers.append(user_int)
        
    else:
        user_int = int(input("Invalid integer or integer already entered, please enter a new 4-digit integer: "))
        if len(str(user_int)) == 4:
            user_numbers.append(user_int)

# This method identifies whether the inputed factors are prime with a sequence of conditions (eg. if the factor is <= 1 
# then the function returns false (not prime), if it is <= 3 than it will return true (prime) and it then checks whether the
# factor is divisible by an increasing sequence of divisors as long as the divisor^2 < the factor, starting from 5, in 
# intervals of 6 up to the the value of the factor itself. If the factor%divisor = 0 then that means that factor has more 
# than two factors and the function returns false. Otherwise if the factor passes all these conditions the function will 
# return True.
def is_prime(n):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# For the range between 2 and the square root of the integer, the integer % count value (i) is calculated and for any iteration
# where this calculation = 0 this count value (i) is added to the factors list. Once this for loop is complete, if the 
# length of the factors list is 2 (2 factors), then each of these factors are passed into the is_prime function to test whether
# they are prime factors. If they are both prime, the has_only_two_prime_factors function returns true, otherwise it returns 
# false.
def has_only_two_prime_factors(num):
    global factors
    factors = []
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            factors.append(i)
            factors.append(num // i)
            break
    if len(factors) == 2 and is_prime(factors[0]) and is_prime(factors[1]):
        return True
    return False

# This method checks whether the integer is a perfect square by checking integer/square root = square root, whether it has only two prime
# factors by checking whether the output from the function has_only_two_prime_factors is true/false or whether the integer has only two largest
# factors which consists of the first factor below the square root where the integer%factor = 0 and the integer//first factor which 
# is the second factor
def closestDivisors():
    for i in user_numbers:
        a = round(math.sqrt(i))
        
        while i%a > 0:
            a -= 1
        
        if i//a == a:
            print(str(i) + " is a perfect square, the square root is " + str(a))
        
        elif has_only_two_prime_factors(i):
            print(i, "has only two prime factors:", str(factors[0]), str(factors[1]))
            
        else:
            print("Largest factors of " + str(i) + " are: " + str(a),str(i//a))

closestDivisors()
```
